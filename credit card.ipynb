```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
data = pd.read_csv('creditcard.csv')
data
```

```python
# info
data.info()
```

```python
# checking the number of missing values in each column

data.isnull().sum()
```

```python
# distribution od legit transactions and fruadulent transaction 

data['Class'].value_counts()
```

# this dataset is highly unbalanced
0 = Normal transaction 
1 = Fraudulent transaction

```python
# separating the data for analysis 

legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions 

data.groupby('Class').mean()
```

```python
# Under-Sampling 
# build a sample dataset containing similiar distribution of normal transactions and fraudlent transactions
# no. of fraudulent transaction is 492 
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatenating two DataFrames

newdata = pd.concat([legit_sample, fraud], axis=0)
```

```python
newdata.head()
```

```python
newdata.tail()
```

```python
newdata['Class'].value_counts()
```

```python
newdata.groupby('Class').mean()
```

```python
# Splitting the data into features and target 
X = newdata.drop(columns='Class',axis=1)
Y = newdata['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# split data into training data and testing data 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
# model training 
# logistic regression 

model = LogisticRegression()
```

```python
# training the logistic regression model with training data 

model.fit(X_train, Y_train)
```

```python
# model evaluation 
# accuracy score

#accuracy on training data 
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on Training data:',training_data_accuracy)
```

```python
# accuracy on test data 
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Testing data:',test_data_accuracy)
```
